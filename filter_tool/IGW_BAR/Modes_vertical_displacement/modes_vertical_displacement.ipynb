Plot for theoretical vertical modes (not used)

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
D = -1000  # Depth of the ocean
a = 100  # Amplitude

# Vertical position (z)
z = np.linspace(0, D, 1000)

# Modes
modes = [1, 2, 3]

# Plot
plt.figure(figsize=(4, 6))

for n in modes:
    Gn = a * np.sin(n * np.pi * z / D)
    plt.plot(Gn, z, label=f'Mode {n}')

plt.title('Vertical Displacement Modes for Internal Tides (IT)')

plt.xlabel('Amplitud (m)')
plt.ylabel('Vertical displacement $G_n(z)$ (m)')
plt.legend()
plt.grid(True)
plt.show()
```
