```python
from PIL import Image
import os

# Define the base path where the folders are located
base_path = '/bettik/PROJECTS/pr-data-ocean/riverama/Frames/BAR_IGW'

# List of folders in the order they should appear in the matrix
folders = [
    'lambda_250', 'lambda_300', 'lambda_350',
    'lambda_400', 'lambda_450', 'lambda_500',
]

# Range of frames
start_frame = 192 # 192
end_frame = 263 # 263

# Create directory for combined images if it does not exist
combined_path = os.path.join(base_path, 'combined_lambdas')
os.makedirs(combined_path, exist_ok=True)

# Iterate over each frame
for frame_num in range(start_frame, end_frame + 1):
    frame_name = f'frame_{frame_num:04d}.png'
    images = []

    # Load the images from each folder
    for folder in folders:
        image_path = os.path.join(base_path, folder, frame_name)
        try:
            with Image.open(image_path) as img:
                images.append(img.copy())
        except IOError:
            print(f"Failed to load image from {image_path}")

    # Combine the images if any were loaded
    if images:
        image_width, image_height = images[0].size
        combined_image = Image.new('RGB', (image_width * 2, image_height * 3))

        # Paste each image into the combined_image
        for index, image in enumerate(images):
            x = index % 2 * image_width
            y = index // 2 * image_height
            combined_image.paste(image, (x, y))

        # Save the combined image
        combined_frame_path = os.path.join(combined_path, frame_name)
        combined_image.save(combined_frame_path)
        print(f"Saved combined image {combined_frame_path}")
    else:
        print(f"No images were loaded for frame {frame_name}. Check your file paths and names.")
```
