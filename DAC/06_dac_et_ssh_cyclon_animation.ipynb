```python
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.gridspec as gridspec
import os

# Base paths
base_path = "/bettik/PROJECTS/pr-data-ocean/riverama/Frames/ssh_et_dac_cyclon_02"
dac_path = os.path.join(base_path, "dac")
ssh_path = os.path.join(base_path, "ssh")
save_path = os.path.join(base_path, "comparison")

# Ensure the save directory exists
os.makedirs(save_path, exist_ok=True)

# Start and end frame numbers (inclusive)
start_frame = 744  # Adjust based on your starting frame number
end_frame = 911    # Adjust based on your ending frame number

# Loop through each frame number
for frame_number in range(start_frame, end_frame + 1):
    # Adjust the file names as needed
    dac_file = os.path.join(dac_path, f"frame_00{frame_number}.png")
    ssh_file = os.path.join(ssh_path, f"frame_00{frame_number}.png")
    
    # Load the images
    dac_image = Image.open(dac_file)
    ssh_image = Image.open(ssh_file)
    
    # Plotting
    fig = plt.figure(figsize=(20, 10))
    gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1])  # Assuming you want equal size, adjust as needed

    ax0 = plt.subplot(gs[0])
    ax0.imshow(dac_image)
    ax0.axis('off')
    
    ax1 = plt.subplot(gs[1])
    ax1.imshow(ssh_image)
    ax1.axis('off')
    
    plt.tight_layout()
    
    # Save the figure
    save_filename = os.path.join(save_path, f"frame_{frame_number - start_frame:04d}.png")
    plt.savefig(save_filename)
    plt.close()  # Close the figure to free memory

print("All frames have been processed and saved.")
```
